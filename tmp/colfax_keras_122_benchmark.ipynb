## Just a single node 

Probably, it is necessary to launch the test twice due to Theano compilation phase before the training

```python
setup_configuration(nodes='1')
```

```python
import time 

cmd = [
    "python",
    os.path.abspath(os.path.join("..", "tmp", "colfax_keras_122_benchmark.py"))
]

process, job_info = submit_job(cmd, 
                               name='colfax_keras_122_benchmark', 
                               cwd='.')

try:
    while True:
        out = process.stdout.readline()    
        if len(out) > 0:        
            print out

        if process.poll() is not None and len(out) == 0:
            break
except KeyboardInterrupt:
    !qdel {job_info['id']}
    time.sleep(1.0)
    !qstat
```
