```python
caffe_help_cmd = [
    "caffe",
    "--help", 
]
process, job_info = submit_job(caffe_help_cmd, name='caffe_help', cwd='/home/u2459/Intel_MobileODT/notebooks/', env='export PATH=$PATH:/opt/caffe-master/build/tools')

while True:
    out = process.stdout.readline()    
    if len(out) > 0:        
        print 'OUT:', out
                
    if process.poll() is not None and len(out) == 0:
        break
```
